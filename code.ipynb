```python
#Gets the rows and columns from specific index
car=sm.iloc[0:11]
car
```

```python
#Bar Plot for Brand and their Range per Kilometer
plt.bar(car['Brand'],car['Model'],color='Green')
plt.xticks(rotation=90)
plt.xlabel("Brand",fontsize=15)
plt.ylabel("Model",fontsize=15)
plt.show()
```

```python

#Bar Plot for Brand and their Range per Kilometer
x=car['Brand']
y=car['Range_Km']
plt.bar(x,y,color='purple')
plt.xticks(rotation=90)
plt.xlabel("Brand",fontsize=15)
plt.ylabel("Range per Km",fontsize=15)
plt.show()
```

```python
#Pie chart showing which Brand has more Top Speed KmH
ab=car['Model']
abc=car['Efficiency_WhKm']
plt.pie(abc,labels=ab,autopct='%0.1f%%')
plt.show()
```

```python
#Pie chart showing which Brand has more Top Speed KmH
br=car['Brand']
sp=car['TopSpeed_KmH']
# c=['magma','gold','red','pink','blue','green','purple','olive','cyan','y']
plt.pie(sp,labels=br,autopct='%0.1f%%')
plt.show()
```

```python
from scipy.stats import skew
# calculate skew
skew(car[['TopSpeed_KmH','Range_Km']],axis=0)
```

```python
from scipy.stats import kurtosis
# calculate Pearson kurtosis
"Pearson Kurtosis:",kurtosis(car[['TopSpeed_KmH','Range_Km']],fisher=False)
```

```python
# calculate Fisher's kurtosis
"Fisher's Kurtosis:",kurtosis(car[['TopSpeed_KmH','Range_Km']],fisher=True)
```

```python
#stack plot
x=car['Brand']
area=car['Range_Km']
area1=car['TopSpeed_KmH']
label=['Range km','Top Speed KmH']
plt.stackplot(x,area,area1,labels=label,colors=['olive','yellow'])
plt.xticks(rotation=90)
plt.legend(loc=2)
plt.show()
```
